```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("weather.csv")
data = data.dropna()
```

```python
sns.jointplot(x=data["MinTemp"], y=data["MaxTemp"]);
```

```python
sns.distplot(data["MaxTemp"])
```

```python
sns.distplot(data["MinTemp"])
```

```python
sns.distplot(data["Rainfall"])
```

```python
sns.kdeplot(data["Evaporation"], shade=True)
```

```python
with sns.axes_style('white'):
    sns.jointplot(x = data['Pressure9am'], y = data['Temp9am'], kind = 'hex', color = 'k')
```

```python
sns.jointplot(x='Cloud9am', y='Humidity9am', data=data, kind='kde')
```

```python
plt.scatter(x=data["WindGustSpeed"], y=data["Sunshine"])
```

```python
data_1=pd.DataFrame(data, columns=['Humidity3pm','Humidity9am','Cloud3pm','Cloud9am','RainToday'])
sns.pairplot(data_1,hue='RainToday')
```

```python
sns.stripplot(x ='Cloud3pm' , y='Humidity3pm', data=data)
```

```python
sns.stripplot(x ='Cloud3pm' , y='Humidity3pm', data=data, jitter=True)
```

```python
sns.swarmplot(x ='Cloud3pm' , y='Humidity3pm', data=data)
```

```python
sns.boxplot(x ='Cloud9am' , y='Humidity9am', data=data)
```

```python
sns.boxplot(x ='Cloud9am' , y='Humidity9am', data=data,hue='RainToday', dodge=True)
```

```python
sns.barplot(x='RainToday', y='Cloud3pm', data=data)
```

```python
sns.barplot(x='Cloud3pm', y='Humidity3pm', data=data, hue='RainToday')
```

```python
sns.countplot(x='WindGustDir', data=data, palette='Greens_d')
```

```python
plt.figure(1)

plt.subplots_adjust(hspace=0.5 )

plt.subplot(211)
sns.countplot(x='RainTomorrow', data=data)

plt.subplot(212)
sns.countplot(x='Cloud3pm', data=data)

plt.show()
```
