```python
#Importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
#Importing DataSets
df=pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/train.csv ")
df1=pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/test.csv")
```

```python
#Concatinating Data Sets
lst=[df,df1]
df=pd.concat(lst)
```

```python
#First five rows of Data Set

df.head()
```

```python
#Drop Null Values from survived column

df=df.dropna(subset=['Survived'])
df.head()
```

```python
#Creating a column that tell us that the passenger has taken a cabin or not
df['Has_Cabin']=df["Cabin"].notnull()
df.head()
```

```python
#creating a column that contains the titles of the passenger like(Mr,Mrs,Master,Miss,[])

def split_it(year):
    return str(re.findall("(Mr|Mrs|Master|Miss)", year))

df["Titles"] = df['Name'].apply(split_it)

df.head()
```

```python
#filling the NAN value in the columns with the mean of their respective columns
df=df.fillna(df.mean())
```

```python
#Creating a column that will tell us the category of the fare.
df['Categorical_fare']=pd.cut(df['Fare'], 3,labels=[ "Economy" ,"Premium-Economy","Business"])
```

```python
#creating A column that will tell us the agge category of the passenger
df['Categorical_age']=pd.cut(df['Age'], [0,13,19,100], labels=["Child", "Teen", "Adult"])
```

```python
df.head()
```

```python
#droping the columns that are not required for the Logistic Regression
df=df.drop(['Cabin','Name','PassengerId','Ticket','Age','Fare','SibSp','Parch'],axis=1)
```

```python
#Converting all variables into numeric ones 
df=pd.get_dummies(df)
```

```python
df.columns
```

```python
#Training and Testing

x=df.drop(["Survived"],axis=1)
y=df['Survived']
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, \
                                                   y, \
                                                   test_size = 0.3, \
                                                   random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
#Prediction

y_pred = classifier.predict(X_test)
```

```python
#Accuracy Score

score = classifier.score(X_test, y_test)
print(score)
```
